```python
# Import Dependencies 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load in csv
life_df = pd.read_csv("LifeExpectancy_2019_Clean.csv")
life_df.head()
```

```python
# x axis and tick locations
x_axis = np.arange(len(life_df))
tick_locations = [value for value in x_axis]
```

```python
# write x labels and set figure size 
plt.figure(figsize=(40,7))
plt.bar(x_axis, life_df["2019 Air Quality Index"], color='purple', alpha=0.75, align="center")
plt.xticks(tick_locations, life_df["Country"], rotation="vertical")

# Set a Title and labels
plt.title("Country & Life Expectancy")
plt.xlabel("Country")
plt.ylabel("2019 Life Expectancy")

# Save graph and show graph
plt.savefig("Country_and_Life_Expenctancy1.png")
```
